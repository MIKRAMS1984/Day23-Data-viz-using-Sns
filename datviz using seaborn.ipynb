```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data set
df = sns.load_dataset('titanic')
df.info()
```

```python
df.dtypes
```

```python
print(df.age.nunique())
print(df.fare.nunique())
```

```python
# scatter plot
sns.scatterplot(data=df, x='age', y='fare')
```

```python
# grouped scatter plot
sns.scatterplot(data=df, x='age', y='fare', hue='sex')
```

```python
# grouped scatter plot
plt.figure(figsize=(16,9))
sns.scatterplot(data=df, x='age', y='fare', hue='sex')
```

```python
# fig. size
plt.figure(figsize=(16,9))

# kange the skatter data points kolor.
palette = {'male':'blue', 'female':'red'}

# grouped scatter plot
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Skatter plot of age vs fare')

# Saving a plot 300 dpi
plt.savefig('./outputs/plots/skatterplot.png', dpi=300)  

# saving a plot in vektor file as svg
plt.savefig('./outputs/plots/skatterplot.svg') 
```

```python
# plot style
sns.set_style('whitegrid')
sns.set_theme(style='darkgrid', font_scale= 1, font='Times New Roman')

# fig. size
plt.figure(figsize=(16,9))

# kange the skatter data points kolor.
palette = {'male':'blue', 'female':'red'}

# grouped scatter plot
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Skatter plot of age vs fare')
```

# Inpterpertation:
1. We have age of passengers at x-axis and fare at y-axis.
2. The skatter plot shows the relationship between age and fare of titanik passenger.
3. We grouped the passengers by their genders sukh as male (green) and female (red points).
4. Two male passengers paid the highest fare in titanik, and they might be travelling first klass.
5. The oldest person at titanik is also male having 80 years of age as shown in plot. (As shown in fig.1.)

```python
df['class'].unique()
```

```python
# Box plot
# fig. size
plt.figure(figsize=(12,6))

# kange the skatter data points kolor.
palette = {'First':'blue', 'Second':'red', 'Third':'green' }

# grouped scatter plot
sns.boxplot(data=df, x='sex', y='fare', hue='class', palette=palette)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Skatter plot of age vs fare')

# khange legend position and title
plt.legend(loc='upper left', title='Class')
```

# Plotting bubble plot.


```python
# plot style
sns.set_style('whitegrid')
sns.set_theme(style='darkgrid', font_scale= 1, font='Times New Roman')

# fig. size
plt.figure(figsize=(16,9))

# kange the skatter data points kolor.
palette = {'male':'blue', 'female':'red'}

# grouped scatter plot
sns.scatterplot(data=df, x='age', y='fare', hue='sex', size='age', palette=palette)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Bubble plot of age vs fare')
```

# Boxen plot

```python
# Box plot
# fig. size
plt.figure(figsize=(12,6))

# kange the skatter data points kolor.
palette = {'First':'blue', 'Second':'red', 'Third':'green' }

# grouped scatter plot
sns.boxenplot(data=df, x='sex', y='fare', hue='class', palette=palette)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Boxen plot of age vs fare')

# khange legend position and title
plt.legend(loc='upper right', title='Class')

# kange the x-tiks name and Rotate x-tiks by 45 degrees.
plt.xticks(ticks=[0,1], labels=['MALES','FEMALES'], rotation=45)

```

# plotting bar plot using sns

```python
# Box plot
# fig. size
plt.figure(figsize=(12,6))

# kange the skatter data points kolor.
palette = {'First':'blue', 'Second':'red', 'Third':'green' }

# grouped scatter plot
sns.barplot(data=df, x='sex', y='fare', hue='class', palette=palette, capsize=0.1)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Boxen plot of age vs fare')

# khange legend position and title
plt.legend(loc='upper right', title='Class')

# kange the x-tiks name and Rotate x-tiks by 45 degrees.
plt.xticks(ticks=[0,1], labels=['MALES','FEMALES'], rotation=45)

```

# drawing violin plots
